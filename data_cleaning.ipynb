## Ejercicio 2: Limpieza de datos de una tienda online

**Escenario real:** Exportaste órdenes de una tienda online y los datos llegaron sucios: precios como texto, fechas mal formateadas y cantidades faltantes.

**Data:** Generada en el ejercicio (simula un CSV exportado de Shopify/WooCommerce)

**Propósito:** Convertir tipos de datos, limpiar columnas de texto y manejar nulos para poder hacer análisis de ventas.

**Resultado esperado:**
- Columna `price` como `float`
- Columna `order_date` como `datetime`
- Columna `quantity` sin nulos (rellenar con 1, porque una orden sin cantidad = 1 unidad)
- Nueva columna `total` = `price * quantity`
- Solo órdenes con `status` = "completed"

```python
import pandas as pd
import numpy as np

orders_data = {
    "order_id": [101, 102, 103, 104, 105, 106],
    "order_date": ["2024-01-15", "2024/01/16", "15-01-2024", "2024-01-18", "2024-01-19", "2024-01-20"],
    "price": ["29.99", "15.50", "abc", "99.00", "45.00", "12.75"],
    "quantity": [2, np.nan, 1, np.nan, 3, 1],
    "status": ["completed", "pending", "completed", "cancelled", "completed", "completed"]
}

orders = pd.DataFrame(orders_data)
print("Datos originales:")
print(orders)
print("\nTipos de datos:")
print(orders.dtypes)

# --- TU SOLUCIÓN AQUÍ ---

# 1. Convertir price a float (hint: usa pd.to_numeric con errors="coerce")
orders["price"] = pd.to_numeric(orders["price"], errors="coerce")

# 2. Convertir order_date a datetime (hint: usa pd.to_datetime con errors="coerce")
orders["order_date"] = pd.to_datetime(orders["order_date"], format="mixed", errors="coerce")

# 3. Rellenar quantity nula con 1
orders["quantity"] = orders["quantity"].fillna(1)

# 4. Crear columna total = price * quantity
orders["total"] = orders["price"] * orders["quantity"]

# 5. Filtrar solo status == "completed"
orders = orders[orders["status"] == "completed"]

# 6. Imprimir resultado final y verificar tipos de datos
print(f"New orders:\n {orders}")
print(f"Type of data: \n {orders.dtypes}")
```

## Ejercicio 3: Combinar datos de empleados y salarios (HR)

**Escenario real:** El área de Recursos Humanos tiene dos tablas separadas: una con info de empleados y otra con sus salarios. Necesitas unirlas y detectar empleados sin salario asignado.

**Data:** Generada en el ejercicio (simula exportaciones de un sistema HR como Workday)

**Propósito:** Practicar `merge` con `left join`, identificar nulos después de unir, y tomar decisiones de limpieza basadas en el contexto del negocio.

**Resultado esperado:**
- DataFrame combinado con todos los empleados
- Empleados sin salario asignado claramente identificados (columna `salary` = `NaN`)
- Nueva columna `salary_category`: "Senior" si salary >= 5000, "Junior" si < 5000, "Sin asignar" si NaN
- Reporte: cuántos empleados no tienen salario

```python
employees = pd.DataFrame({
    "emp_id": [1, 2, 3, 4, 5, 6],
    "name": ["Laura", "Carlos", "Mia", "Jorge", "Sofia", "Pedro"],
    "department": ["Engineering", "Marketing", "Engineering", "HR", "Marketing", "Engineering"],
    "hire_date": ["2021-03-01", "2020-07-15", "2022-01-10", "2019-05-20", "2023-08-01", "2021-11-30"]
})

salaries = pd.DataFrame({
    "emp_id": [1, 2, 3, 5],
    "salary": [7000, 4500, 8500, 4200]
})

print("Empleados:")
print(employees)
print("\nSalarios:")
print(salaries)

# --- TU SOLUCIÓN AQUÍ ---

# 1. Hacer left join de employees con salaries usando emp_id
df_combined = pd.merge(employees,salaries, on="emp_id",how="left")

# 2. Crear columna salary_category:
#    - "Senior" si salary >= 5000
#    - "Junior" si salary < 5000
#    - "Sin asignar" si NaN
#    (hint: usa np.select o pd.cut, o una función lambda con apply)
df_combined["salary_category"] = df_combined["salary"].apply(
    lambda x: "Sin asignar" if pd.isna(x) else ("Senior" if x >= 5000 else "Junior")
)


# 3. Convertir hire_date a datetime
df_combined["hire_date"] = pd.to_datetime(df_combined["hire_date"],format="mixed", errors="coerce")

# 4. Imprimir cuántos empleados no tienen salario asignado
print(f"Employee dont have salary assigned: \n {df_combined["salary"].isna().sum()}")

# 5. Imprimir el DataFrame final ordenado por salary descendente
print(df_combined.sort_values(by="salary", ascending=False))
```

## Ejercicio 4: Precios de acciones con datos faltantes (Series de tiempo)

**Escenario real:** Descargaste precios de cierre de una acción y faltan días (fines de semana, feriados, o errores en la API). Necesitas reconstruir la serie completa.

**Data:** Generada en el ejercicio (simula datos de Yahoo Finance / Alpha Vantage)

**Propósito:** Aprender cuándo usar `ffill`, `bfill` e `interpolate` en el contexto correcto. En finanzas, `ffill` es el estándar porque el precio "no cambió" si no hubo trading.

**Resultado esperado:**
- Serie de fechas completa (sin saltos)
- Nulos rellenados con `ffill` (precio del día anterior)
- Columna adicional `daily_change` = diferencia entre precio de hoy y el anterior
- Identificar el día con mayor subida y mayor bajada

```python
dates = pd.date_range(start="2024-01-01", end="2024-01-15")
prices = [150.0, np.nan, np.nan, 152.5, 153.0, np.nan, 154.2, 155.0, np.nan, 153.8, 152.0, np.nan, np.nan, 156.5, 158.0]

stocks = pd.DataFrame({
    "date": dates,
    "close_price": prices
})

print("Serie original (con nulos):")
print(stocks)
print(f"\nTotal de nulos: {stocks['close_price'].isna().sum()}")

# --- TU SOLUCIÓN AQUÍ ---

# 1. Establecer 'date' como índice del DataFrame
stocks = stocks.set_index('date')

# 2. Rellenar nulos con forward fill (ffill)
stocks["close_price"] = stocks["close_price"].ffill()

# 3. Crear columna 'daily_change' = diferencia con el día anterior
#    (hint: usa .diff())
stocks["daily_change"] = stocks["close_price"].diff()

# 4. Imprimir el día con mayor subida (max de daily_change)
print(f"Day most up: \n {stocks["daily_change"].idxmax()}")

# 5. Imprimir el día con mayor bajada (min de daily_change)
print(f"Day most up: \n {stocks["daily_change"].idxmin()}")

# 6. Imprimir la serie completa limpia
print(f"Stocks: {stocks}")
```

## Ejercicio 5: Base de datos de clientes con duplicados y datos inconsistentes

**Escenario real:** Importaste una base de clientes de un CRM y tiene registros duplicados, emails con mayúsculas/minúsculas mezcladas, espacios extra en nombres, y edades imposibles (errores de digitación).

**Data:** Generada en el ejercicio (simula exportación de Salesforce / HubSpot)

**Propósito:** Aprender a detectar y eliminar duplicados, estandarizar texto, y filtrar outliers basados en reglas de negocio.

**Resultado esperado:**
- Sin registros duplicados (mismo email = mismo cliente)
- Emails en minúsculas y sin espacios
- Nombres sin espacios al inicio/final y con capitalización correcta
- Edades inválidas (menores de 0 o mayores de 120) reemplazadas por `NaN`
- Reporte final: cuántos duplicados se eliminaron y cuántas edades inválidas se encontraron

```python
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5, 6, 7],
    "name": ["  juan perez", "MARIA LOPEZ", "Carlos Ruiz", "juan perez  ", "ANA GARCIA", "  Luis Torres", "Maria Lopez"],
    "email": ["JUAN@EMAIL.COM", "maria@email.com", "carlos@email.com", "juan@email.com", "ana@email.com", "luis@email.com ", "MARIA@EMAIL.COM"],
    "age": [28, 34, -5, 28, 999, 41, 34],
    "city": ["Bogotá", "Medellín", "Cali", "Bogotá", "Barranquilla", "Cali", "Medellín"]
})

print("Clientes originales:")
print(customers)
print(f"\nTotal de registros: {len(customers)}")

# --- TU SOLUCIÓN AQUÍ ---

# 1. Limpiar columna 'email': minúsculas y sin espacios
#    (hint: .str.lower() y .str.strip())
customers["email"] = customers["email"].str.lower().str.strip()

# 2. Limpiar columna 'name': sin espacios al inicio/final y capitalización correcta
#    (hint: .str.strip() y .str.title())
customers["name"] = customers["name"].str.strip().str.title()

# 3. Reemplazar edades inválidas (< 0 o > 120) con NaN
#    (hint: usa np.where o mask)
customers["age"] = np.where((customers['age'] < 0) | (customers['age'] > 120), np.nan, customers['age'])

# 4. Eliminar duplicados basándose en la columna 'email'
#    (hint: .drop_duplicates(subset="email"))
customers = customers.drop_duplicates(subset="email")

# 5. Imprimir cuántos duplicados se eliminaron y cuántas edades inválidas había
print(f"Show duplicates: \n {customers["email"].duplicated().sum()}")

# 6. Imprimir el DataFrame limpio final
print(f"Clean dataframe: \n {customers}")
```

## Ejercicio 6: Consolidar ventas de dos sucursales

**Escenario real:** Tu empresa tiene dos sucursales que guardan sus ventas en archivos separados con nombres de columnas distintos. Necesitas unirlos en un solo reporte mensual.

**Data:** Generada en el ejercicio (simula reportes de ventas de dos tiendas)

**Propósito:** Practicar `concat`, `rename`, y generar un reporte agregado con `groupby`. Este flujo es muy común en empresas con múltiples puntos de venta.

**Resultado esperado:**
- Un solo DataFrame con ventas de ambas sucursales
- Columnas estandarizadas: `date`, `product`, `quantity`, `revenue`, `branch`
- Reporte de ventas totales por producto (suma de revenue)
- Producto más vendido por cada sucursal

```python
branch_north = pd.DataFrame({
    "fecha": ["2024-01-01", "2024-01-01", "2024-01-02", "2024-01-02"],
    "producto": ["Laptop", "Mouse", "Laptop", "Teclado"],
    "cant": [3, 10, 2, 5],
    "ingresos": [3000, 150, 2000, 200]
})

branch_south = pd.DataFrame({
    "sale_date": ["2024-01-01", "2024-01-01", "2024-01-02", "2024-01-02"],
    "item": ["Mouse", "Teclado", "Laptop", "Mouse"],
    "units_sold": [8, 4, 1, 12],
    "total_sales": [120, 160, 1000, 180]
})

print("Sucursal Norte:")
print(branch_north)
print("\nSucursal Sur:")
print(branch_south)

# --- TU SOLUCIÓN AQUÍ ---

# 1. Renombrar columnas de ambos DataFrames al estándar:
#    date, product, quantity, revenue
branch_north = branch_north.rename(columns={"fecha":"date",
                                            "producto":"product",
                                            "cant":"quantity",
                                            "ingresos":"revenue"})

branch_south = branch_south.rename(columns={"sale_date":"date",
                                            "item":"product",
                                            "units_sold":"quantity",
                                            "total_sales":"revenue"})

# 2. Agregar columna 'branch' a cada uno: "Norte" y "Sur"
branch_south["branch"] = "Norte"
branch_north["branch"] = "Sur"

# 3. Concatenar ambos DataFrames (axis=0)
combined_branch = pd.concat([branch_north,branch_south], axis=0,ignore_index=True)

# 4. Convertir 'date' a datetime
combined_branch["date"] = pd.to_datetime(combined_branch["date"], errors="coerce")

# 5. Reporte: revenue total por producto (groupby + sum)
print(f"Report revenue total por producto: \n: {combined_branch.groupby("product")["revenue"].sum()}")

# 6. Producto con mayor revenue por sucursal (groupby + idxmax)
print(f"Report con mayor revenue por sucursal: \n: {combined_branch.groupby("branch")["revenue"].idxmax()}")
```
