# Basic Dataframe operations

```python
# Viewing data
df.head()  # First 5 rows
df.tail()  # Last 5 rows
df.info()  # Summary of the DataFrame
df.describe()  # Statistical summary of numerical columns
```

```python
# Selection and indexing
print(df['Name'])  # Select a single column
print(df[['Name', 'Age']])  # Select multiple columns
print(df.iloc[0])  # Select the first row
print(df.loc[0, 'Name'])  # Select a specific value
```

```python
#Filter rows
print(df[df['Age'] > 30])  # Filter rows where Age is greater than 30
```

```python
# Selection by position
print(df.iloc[0])  # Select the first row
print(df.iloc[0:2])  # Select the first two rows
print(df.iloc[:, 0])  # Select the first column
print(df.iloc[:, 0:2])  # Select the first two columns
```

```python
# Selection by label
print(df.loc[0])  # Select the first row
print(df.loc[0:1])  # Select the first two rows
print(df.loc[:, 'Name'])  # Select the 'Name' column
```
