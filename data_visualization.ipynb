# Learn Data visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.show()
```

## Basic Plot

```python
# Linear plot
plt.plot([1,2,3], [10,20,30], label="Trend", color="orange", linestyle="--", marker="o")
plt.title("Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend()
plt.show()
```

```python
# Bar Chart
categories = ["A", "B", "C"]
values = [10,15,7]
plt.bar(categories,values, color="blue")
plt.title("Bar Chart")
plt.show()
```

```python
# Histogram
data = [1,2,2,4,4,4,3,3]
plt.hist(data, bins=4, color="green", edgecolor="black")
plt.title("Histogram")
plt.show()
```

```python
#Scatter Plot
x = [1,2,3,4,5]
y = [10,12,25,30,45]
plt.scatter(x,y,color="red")
plt.title("Scatter Plot")
plt.show()
```

## Customizing Plots

```python
x = [1,2,3,4,5]
y = [10,12,25,30,45]
plt.scatter(x,y,color="red")
plt.title("Scatter Plot")
plt.xlabel("X-axis Label")
plt.xlabel("Y-axis Label")
plt.legend(["Dataset1"])
plt.show()
```
