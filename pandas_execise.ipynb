## Bloq 5 - Null values

```python
data_null = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, None, 40, None],
    "Note": [85, 90, 78, None, 88]
}
df = pd.DataFrame(data_null)
```

```python
#Count the number of null values in each column
print(f"Number of null values in each column: \n{df.isnull().sum()}")
```

```python
df.loc[df["Age"].isnull(), "Age"] = df["Age"].mean()
print(f"Data with null values in Age column filled with mean: \n{df}")
```

## Bloq 6 - Groupby and agreggation

### Exercise 10

```python
new_data = {
    "product": ["A", "B", "C", "D", "E"],
    "price": [10, 20, 15, 25, 30],
    "city": ["New York", "New York", "Chicago", "Chicago", "Chicago"]
}

df_new = pd.DataFrame(new_data)
```

```python
# Calculate the total to city
result = df_new.groupby("city")["price"].sum()
print(f"Total price by city: \n{result}")
```

```python
# Get product with most bigger price
max_price_product = df_new.loc[df_new["price"].idxmax()]
print(f"Product with the biggest price: \n{max_price_product}")
```
