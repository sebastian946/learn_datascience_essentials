# Numpy learning course

```python
# Import numpy packages
import numpy as np
```

```python
#Creating array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
# Building functions
zeroes = np.zeros((3, 3))
print(zeroes)
```

```python
ones = np.ones((2, 4))
print(ones)
```

```python
# Using range
range_array = np.arange(1, 10, 2)
print(range_array)
```

```python
# Linkspace
linspace_array = np.linspace(0, 1, 5)
print(linspace_array)
```

```python
# Manipulating arrays
arr = np.array([1, 2, 3, 4, 5, 6])
reshape = arr.reshape((2, 3))
print(reshape)
```

```python
# Expanded
arr = np.array([1, 2, 3])
expandede = arr[:, np.newaxis]
print(expandede)
```

# Array Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(a + b)
print(a * b)
print(a / b)
```

```python
#Mathematical operations
arr = np.array([4, 16, 25])
print(np.sqrt(arr))
print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
```

# Array indexing, Slicing and Reshaping

```python
# Indexing and slicing
arr = np.array([10, 20, 30, 40, 50, 60])
print(arr[0])
print(arr[-1])
print(arr[1:4])
print(arr[:3])
print(arr[3:5])

#Reshape

reshaped = arr.reshape(2,3)
print(reshaped)
```

# Exercise 2

```python
# Create array 3x3 and perform operations
matrix = np.arange(1, 10).reshape(3, 3)
print(f"Oringinal Matrix:\n{matrix}")

#Transpose
transpose = matrix.T
print(f"Transposed Matrix:\n{transpose}")

another_matrix = np.arange(10, 19).reshape(3, 3)
print(f"Another Matrix:\n{another_matrix}")
print(f"Matrix Addition:\n{matrix + another_matrix}")
print(f"Matrix Multiplication:\n{matrix * another_matrix}")
```

```python
c = np.arange(1,17).reshape(4,4)
print(f"Sum of all elements: {np.sum(c)}")
```

```python
# Exercise
temperatures = np.array([18, 22, 35, 29, 31, 27, 24])
print(f"Normalized Temperatures: {(temperatures - np.min(temperatures)) / (np.max(temperatures) - np.min(temperatures))}")
print(f"The most hot day was: {temperatures[np.argmax(temperatures)]} degrees")
print(f"The day with the hot temperature was: {np.max(temperatures)} degrees")
```

```python
arr_4 = np.arange(1, 17).reshape(4, 4)
print(f"Mean of each column: {np.mean(arr_4, axis=0)}")
min = np.min(arr_4)
max = np.max(arr_4)
normalized_arr_4 = (arr_4 - min) / (max - min)
print(f"Normalized Array:\n{normalized_arr_4}")
```
