# Numpy learning course

```python
# Import numpy packages
import numpy as np
```

```python
#Creating array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
```

```python
# Building functions
zeroes = np.zeros((3, 3))
print(zeroes)
```

```python
ones = np.ones((2, 4))
print(ones)
```

```python
# Using range
range_array = np.arange(1, 10, 2)
print(range_array)
```

```python
# Linkspace
linspace_array = np.linspace(0, 1, 5)
print(linspace_array)
```

```python
# Manipulating arrays
arr = np.array([1, 2, 3, 4, 5,6])
reshape = arr.reshape((2, 3))
print(reshape)
```
